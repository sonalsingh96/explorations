# Resume Parsing

###### import libraries

```python
import glob
import pandas as pd
from resume_parser import resumeparse 
```

```python
list_of_resume = glob.glob('/Users/sonalsingh/Downloads/Examples/*.pdf')
```

```python
parsed_df = pd.DataFrame(columns= ['email', 'phone', 'name', 'total_exp', 'university', 'designition', 'degree', 'skills', 'Companies worked at'])
```

###### Parsing each resume in the file 

```python
for each_resume in list_of_resume:
    data = resumeparse.read_file(each_resume)
    data_unlist = {k: str(v) for k,v in data.items()}
    print(each_resume)
    parsed_df = parsed_df.append(pd.DataFrame(data_unlist,index =[0]))
```
